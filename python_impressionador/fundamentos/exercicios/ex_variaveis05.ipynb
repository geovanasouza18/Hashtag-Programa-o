#### 3. Faça um Programa que peça dois números e imprima a soma.

```python
num1 = int(input('Digite o primeiro número: '))
num2 = int(input('Digite o segundo número: '))
soma = num1 + num2
print(f'A soma dos números foi {soma}')
```

#### 4. Faça um Programa que peça as 4 notas bimestrais de um aluno e mostre a média de todas as notas.

```python
nota1, nota2, nota3, nota4 = map(float, input('Digite as 4 notas dos alunos: ').split())
media = (nota1 + nota2 + nota3 + nota4) / 4
print(f'A média do aluno foi {media:.2f}')
```

#### 5. Faça um Programa que converta metros para centímetros. Você pode pedir o comprimento em metros para o usuário (input).

```python
metros = int(input('Digite a quantidade de metros: '))
centimetros = metros * 100
print(f'A conversão de metros para centímetros foi {centimetros}')
```

#### 6. Faça um Programa que calcule a área de uma sala de um apartamento. Para isso, o seu programa precisa pedir a largura da sala, o comprimento da sala e imprimir a área em m² da sala.

```python
largura, comprimento = map(float, input('Digite a largura e o comprimento da sala: ').split())
area = comprimento * largura
print(f'A área da sala é {area:.2f}m\u00b2')
```

#### 7. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês.

```python
valor_hora = float(input('Qunato você ganha por hora? '))
horas_trabalhadas = int(input('Digite suas horas trabalhadas no mês: '))
salario = valor_hora * horas_trabalhadas
print(f'O seu salário do mês é {salario:.2f}')
```

#### 8. Vamos criar um conversor de temperatura. Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
$C = \frac{5}{9}(F-32)$

```python
fahrenheit = float(input('Digite uma temperatura em fahrenheit: '))
celsius = (5 / 9 * (fahrenheit - 32))
print(f'A temperatura em Celsius é {celsius:.2f}')
```

#### 9. Faça um Programa que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.
$F = \frac{9}{5}C + 32$

```python
celsius = float(input('Digite uma temperatura em Celsius: '))
fahrenheit = (5 / 9 * celsius) + 32
print(f'A temperatura em Fahrenheit é {fahrenheit:.2f}')
```

#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
valor_hora2 = float(input('Quanto você ganha por hora? '))
horas_trabalhadas2 = int(input('Digite suas horas trabalhadas no mês: '))
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
salario2 = valor_hora2 * horas_trabalhadas2
print(salario2)
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
descontoir = (salario2 * 0.11)
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
inss = (salario2 * 0.08)
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
sindicato = (salario2 * 0.05)
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
descontos = salario2 - (descontoir + inss + sindicato)
print(f'O salário com todos os descontos é {descontos:.2f}')
```
